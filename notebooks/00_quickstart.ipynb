```python
from otxlearner.datasets import load_ihdp
from otxlearner.trainers import SinkhornTrainer
train, val, test = load_ihdp(split='all', device='cpu')
trainer = SinkhornTrainer(lambda_max=10.0, epsilon=0.1, device='cpu')
hist = trainer.fit(train, val, epochs=3)
```

```python
import matplotlib.pyplot as plt
plt.plot(hist['epoch'], hist['pehe_val'])
plt.xlabel('epoch'); plt.ylabel('PEHE')
```

```python
print('Final PEHE:', hist['pehe_val'][-1])
```
