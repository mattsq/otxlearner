```python
from otxlearner.data import load_ihdp, torchify
from otxlearner.trainers import SinkhornTrainer
from otxlearner.loops import prepare_loaders
ds = torchify(load_ihdp())
train_loader, val_loader = prepare_loaders(ds, batch_size=512)
trainer = SinkhornTrainer(ds.train.x.shape[1], lambda_max=10.0, epsilon=0.1, device='cpu')
hist = trainer.fit(train_loader, val_loader, epochs=3)
```

```python
import matplotlib.pyplot as plt
plt.plot(hist['epoch'], hist['pehe_val'])
plt.xlabel('epoch'); plt.ylabel('PEHE')
```

```python
print('Final PEHE:', hist['pehe_val'][-1])
```
